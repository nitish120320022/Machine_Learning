```python
#Check feature is numerical or catagorical

numerical_features = [feature for feature in df.columns if df[feature].dtype != 'O']
catagorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']
print('Numerical Features:', numerical_features)
print('-------------------------------------------------------------------------------')
print('Catagorical Features:', catagorical_features)
```

```python
#Check special character

for i in df.columns:
    print(df[i].loc[df[i] == '?'].count())
```

```python
#Replace special characters

rem = df['normalized-losses'].loc[df['normalized-losses'] != '?']
rmedian = rem.astype(str).astype(int).median()
df['normalized-losses'] = df['normalized-losses'].replace('?', rmedian).astype(int)

rem = df['price'].loc[df['price'] != '?']
rmedian = rem.astype(str).astype(int).median()
df['price'] = df['price'].replace('?', rmedian).astype(int)

rem = df['peak-rpm'].loc[df['peak-rpm'] != '?']
rmedian = rem.astype(str).astype(int).median()
df['peak-rpm'] = df['peak-rpm'].replace('?', rmedian).astype(int)

rem = df['horsepower'].loc[df['horsepower'] != '?']
rmedian = rem.astype(str).astype(int).median()
df['horsepower'] = df['horsepower'].replace('?', rmedian).astype(int)

rem = df['stroke'].loc[df['stroke'] != '?']
rmedian = rem.astype(float).median()
df['stroke'] = df['stroke'].replace('?', rmedian).astype(float)

rem = df['bore'].loc[df['bore'] != '?']
rmedian = rem.astype(float).median()
df['bore'] = df['bore'].replace('?', rmedian).astype(float)

rem = df['num-of-doors'].loc[df['num-of-doors'] != '?']
rmedian = rem.mode()[0]
df['num-of-doors'] = df['num-of-doors'].replace('?', rmedian)
```
